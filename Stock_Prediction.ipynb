```
#Get the daily percentage change
Apple['returns'] = (Apple['Close'] / Apple['Close'].shift(1))-1
```

```
Apple.head
```

```
Google['returns'] = (Google['Close'] / Google['Close'].shift(1))-1
```

```
Google.head
```

```
Samsung['returns'] = (Samsung['Close']/Samsung['Close'].shift(1))-1
```

```
Samsung.head
```

```
#Get the Returns
Apple['returns'].hist(bins=100, label='Apple', alpha=0.5, figsize=(15,9))
Google['returns'].hist(bins=100, label='Google', alpha=0.5)
Samsung['returns'].hist(bins=100, label='Samsung', alpha=0.5)
plt.legend()
```

```
#Get the Cummulative Returns
Apple['Cummulative Returns'] = 1 + Apple['returns'].cumprod()
```

```
Apple.head
```

```
Google['Cummulative Returns'] = 1 + Google['returns'].cumprod()
```

```
Google.head
```

```
Samsung['Cummulative Returns'] = 1 + Samsung['returns'].cumprod()
```

```
Samsung.head
```

```
#plot the Cummulative Returns
Apple['Cummulative Returns'].plot(label='Apple', figsize=(13,6))
Google['Cummulative Returns'].plot(label='Google')
Samsung['Cummulative Returns'].plot(label='Samsung')
plt.legend()
```
